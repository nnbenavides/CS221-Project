```python
# get the data
data2 = pd.read_csv("data/flights.csv")
data2 = data2[np.isfinite(data2['ARRIVAL_DELAY'])]
# data = data2.sample(10000)
print(list(data2.columns.values))
```

```python
# start limiting the data
relevant_cols = data2[['ARRIVAL_DELAY', 'DAY','MONTH', 'DAY_OF_WEEK', 'AIRLINE','ORIGIN_AIRPORT', 'DESTINATION_AIRPORT','SCHEDULED_DEPARTURE', 'TAIL_NUMBER']]
relevant_cols = relevant_cols[relevant_cols['ORIGIN_AIRPORT'] == 'ORD']
relevant_cols = relevant_cols[relevant_cols['AIRLINE'].isin(['MQ', 'UA', 'AA', 'EV', 'OO'])]
relevant_cols.head()
```

```python
# make the arrival delay buckets
for row in range(relevant_cols.shape[0]):
    if row % 1000 == 0:
        print ('done ' + str(row))
    delay = relevant_cols.iloc[row, 0]
    if delay < 0:
        relevant_cols.iloc[row, 0] = 0
    elif delay <= 15:
        relevant_cols.iloc[row, 0] = 1
    else:
        relevant_cols.iloc[row, 0] = 2
    # bucket the departure by hour
    relevant_cols.iloc[row, 7] = int(relevant_cols.iloc[row, 7]) / 100
```

```python
# get airline info
data2 = pd.read_csv("data/flights-limited.csv")
data2 = data2[np.isfinite(data2['ARRIVAL_DELAY'])]
relevant_cols = data2[['ARRIVAL_DELAY', 'MONTH', 'DAY_OF_WEEK', 'AIRLINE','ORIGIN_AIRPORT', 'DESTINATION_AIRPORT','SCHEDULED_DEPARTURE', 'TAIL_NUMBER']]

# sample so that the clases are equal
early = relevant_cols[relevant_cols['ARRIVAL_DELAY'] < 0]
early = early.sample(48000)
on_time = relevant_cols[relevant_cols['ARRIVAL_DELAY'] > 0]
on_time = on_time[on_time['ARRIVAL_DELAY'] < 16]
on_time = on_time.sample(48000)
delayed = relevant_cols[relevant_cols['ARRIVAL_DELAY'] > 15]
delayed = delayed.sample(48000)

flights_df = pd.concat([early, on_time, delayed])
```

```python
# need to join the airlane data w/ the flight data
airplane_df = pd.read_csv('data/airplane.csv')
bayes_data_full = pd.merge(airplane_df, flights_df, how='inner', left_on='TAIL_NUMBER', right_on='TAIL_NUMBER')
bayes_data_full = bayes_data_full.drop(columns=['TAIL_NUMBER'])

# split into test and train
bayes_data, test_data = train_test_split(bayes_data_full, test_size=0.2)
test_labels = test_data[['ARRIVAL_DELAY']]
test_data = test_data.drop(columns=['ARRIVAL_DELAY'])
```

```python
# CONSTANTS
NUM_TIME_DELAY_BUCKETS = 5
NUM_AIRPLANE_DELAY_BUCKETS = 5
la_place_constant = 1

# map the variables to their parents
variables_to_parents = {}
variables_to_parents['MONTH'] = []
variables_to_parents['DAY_OF_WEEK'] = []
variables_to_parents['SCHEDULED_DEPARTURE'] = []
variables_to_parents['time_delay'] = ['MONTH', 'DAY_OF_WEEK', 'SCHEDULED_DEPARTURE']
variables_to_parents['AIRLINE'] = []
variables_to_parents['DESTINATION_AIRPORT'] = []
variables_to_parents['YEAR'] = [] # year is the year of the airplane make NOT the year that the flight departs
variables_to_parents['MODEL'] = []
variables_to_parents['airplane_delay'] = ['YEAR', 'MODEL']
variables_to_parents['ARRIVAL_DELAY'] = ['time_delay', 'AIRLINE', 'DESTINATION_AIRPORT', 'airplane_delay']
```

```python
# initialize the counts
variable_to_counts = {}
# these variables don't have parent variables
for var in variables_to_parents:
    if len(variables_to_parents[var]) == 0:
        variable_to_counts[var] = {key:la_place_constant for key in np.unique(bayes_data[var].astype(str).unique())}
# add the variable with parent variables
variable_to_counts['time_delay'] = {(time_delay, month, day_of_week, sch_departure):la_place_constant
                                   for time_delay in range(NUM_TIME_DELAY_BUCKETS)
                                   for month in bayes_data['MONTH'].astype(str).unique()
                                   for day_of_week in bayes_data['DAY_OF_WEEK'].astype(str).unique()
                                   for sch_departure in bayes_data['SCHEDULED_DEPARTURE'].astype(str).unique()}
variable_to_counts['airplane_delay'] = {(airplane_delay, year, model):la_place_constant
                                       for airplane_delay in range(NUM_AIRPLANE_DELAY_BUCKETS)
                                       for year in bayes_data['YEAR'].astype(str).unique()
                                       for model in bayes_data['MODEL'].astype(str).unique()}
variable_to_counts['ARRIVAL_DELAY'] = {(arrival_delay, time_delay, airline, dest_airport, airplane_delay):la_place_constant
                                      for arrival_delay in bayes_data['ARRIVAL_DELAY'].astype(str).unique()
                                      for time_delay in range(NUM_TIME_DELAY_BUCKETS)
                                      for dest_airport in bayes_data['DESTINATION_AIRPORT'].astype(str).unique()
                                      for airplane_delay in range(NUM_AIRPLANE_DELAY_BUCKETS)
                                      for airline in bayes_data['AIRLINE'].astype(str).unique()}
```

```python
# learn a basic Bayesian network (flat structure)
# count
for row in bayes_data:
    for (variable, parent_list) in variables_to_parents.items():
        # need to make tuple of value with their parent(conditional) values
        value = [row[variable]]
        value.extend([row[parent] for parent in parent_list])
        value = tuple(value)
        # increase the count
        variable_to_counts[var][value] += 1
```

```python
# normalize
variable_to_probability = {}
for variable in variable_to_counts:
    if len(variable_to_parents[variable]) > 0:
        # then we need to do some fun tuple stuff because the conditionals are what is constant
        conditional_to_count = {}
        for value in variable_to_counts[variable]:
            conditional = value[1:]
            conditional_to_count[conditional] = conditional_to_count.get(conditional_to_count, 0) + 1
        # now actually get the percentage
        for value in variable_to_counts[variable]:
            conditional = value[1:]
            variable_to_probability[variable][value] = float(variable_to_counts[variable][value])/conditional_to_count[conditional]
    else:
        # there are no parents
        for value in variable_to_counts[variable]:
            # add all of the counts for all of the values for the variable
            total = sum(variable_to_counts[variable][value].values())
            variable_to_probability[variable][value] = float(variable_to_counts[variable][value])/total
```
